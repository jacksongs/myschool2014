```python
import pandas as pd
import numpy as np
import seaborn
%pylab inline

df = pd.read_csv('2014.csv')
df.ix[:,21:-1] = df.ix[:,21:-1].replace("-",np.NaN)
df['Full-time equivalent enrolments'] = df['Full-time equivalent enrolments'].replace("-",np.NaN)
df['Full-time equivalent enrolments'] = df['Full-time equivalent enrolments'].astype(np.float64)
df.ix[:,21:-1] = df.ix[:,21:-1].astype(np.float64)
df['Year 5 | Grammar'].astype(np.float64).hist(bins=df['Year 5 | Grammar'].max()/20)
```

```python
#Alright that's working nicely. Now let's make a big dataframe with all the scores.
fourt = df[df['Year']==2014]
unpacked = pd.melt(fourt.ix[:,1:],id_vars=['Average ICSEA value',
                        'Boys',
                        'Completed senior secondary school',
                        'Full-time equivalent enrolments',
                        'Full-time equivalent non-teaching staff',
                        'Full-time equivalent teaching staff',
                        'Girls',
                        'Indigenous students',
                        'Location',
                        'Name',
                        'Non-teaching staff',
                        'SMCLID',
                        'School ICSEA value',
                        'School sector',
                        'School type',
                        'School-based apprenticeships and traineeships', 
                        'Senior secondary certificate awarded', u'Teaching staff',
                        'Total enrolments',
                        'VET enrolments',
                        'Year',
                        'Year range'])
years = ['3','5','7','9']
domains = ['Grammar','Reading','Numeracy','Writing','Spelling']
for y in years:
    unpacked[y] = unpacked['variable'].str.contains(y)
for d in domains:
    unpacked[d] = unpacked['variable'].str.contains(d)
unpacked['Class'] = unpacked['variable'].str[5].apply(lambda x: int(x))
unpacked['Domain'] = unpacked['variable'].str[9:]

def colo(x):
    if x == 'Government':
        return 1
    else:
        return 2

unpacked['Colour'] = unpacked['School sector'].apply(lambda x: colo(x))

def remoto(x):
    if x == 'Metropolitan':
        return 1
    elif x == 'Provincial':
        return 2
    elif x == 'Remote':
        return 3
    elif x == 'Very Remote':
        return 4
    
unpacked['Remoteness'] = unpacked['Location'].apply(lambda x: remoto(x))
```

```python
new_columns = unpacked.columns.values
new_columns[-14] = 'NAPLAN Score'
unpacked.columns= new_columns
```
